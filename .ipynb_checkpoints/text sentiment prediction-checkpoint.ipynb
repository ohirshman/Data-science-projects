```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB 
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import preprocessing
```

```python
reviews = pd.read_csv("C:\\Users\\or.hirshman\\Documents\\\learning\\data_mining\\hw\\final_project\\hw\\text_training.csv")
```

```python
y = reviews['rating'].values
```

```python
features = reviews[reviews.columns[1:-1]]
```

```python
logisticReg = LogisticRegression(random_state=0, max_iter=1000)
```

```python
rfecv = RFECV(estimator=logisticReg, cv=7, scoring='accuracy')
```

```python
rfecv = rfecv.fit(features,y)
```

```python
print (rfecv.support_)
```
