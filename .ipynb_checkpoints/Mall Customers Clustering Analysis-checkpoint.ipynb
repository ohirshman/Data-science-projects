```python
#Creating dummy variable
gender_dummy = pd.get_dummies(data['Gender'])

#Concatenating the dummy variable to the original dataset
data_dummy = pd.concat([data, gender_dummy], axis=1)

#Deleting categorical variable from the dummy set
del data_dummy['Gender']

data_dummy.head()
```

```python
# Scale  data using StandardScaler 

scaler = StandardScaler()
x_scaled = scaler.fit_transform(data_dummy)
scaled_data = pd.DataFrame(x_scaled, columns=data_dummy.columns)
```

```python
scaled_data.head(3)
```

```python
# K means iterations
wcss=[]
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10, random_state=0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
    
```

```python
# Plot the elbow graph
plt.plot(range(1,11),wcss)
plt.title('Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('inertias')
plt.show()
```

```python
# K means 

kmeans = KMeans(n_clusters=2, random_state=0).fit(scaled_data)
identified_clusters = kmeans.fit_predict(scaled_data)
```

```python
#Appending the identified clusters to the dataframe
clustered_data = data.copy()
clustered_data['Cluster'] = identified_clusters
display(clustered_data.sort_values(by='Cluster'))
```

```python
pca = PCA(n_components=2)
principalComponentsWithGender = pca.fit_transform(scaled_data)
principalDfWithGender = pd.DataFrame(data = principalComponentsWithGender
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDfWithGender.head(3)
```

```python
finalDfWithGender = pd.concat([principalDfWithGender, clustered_data[['Cluster']]], axis = 1)
finalDfWithGender.head(3)
```

```python
# Plot k means results with Gender
fig = px.scatter(principalComponentsWithGender, x=0, y=1, color=finalDfWithGender['Cluster'])
fig.show()
```

```python
# explained variance = how much information (variance) can be attributed to each of the principal components
sum_pca = pca.explained_variance_ratio_[0] + pca.explained_variance_ratio_[1]
sum_pca
```

```python
# Separate the data by Gender and make K Means analysis for each gender
male_data = data.loc[data['Gender']=='Male']
female_data = data.loc[data['Gender']=='Female']

del male_data['Gender']
del female_data['Gender']
```

```python
male_data.head(3)
```

```python
female_data.head(3)
```

```python
# scale male_data and female_data
male_scaled = scaler.fit_transform(male_data)
male_scaled_data = pd.DataFrame(male_scaled, columns=male_data.columns)

female_scaled = scaler.fit_transform(female_data)
female_scaled_data = pd.DataFrame(female_scaled, columns=female_data.columns)
```

```python
male_scaled_data.head(3)
```

```python
female_scaled_data.head(3)
```

```python
# Male KMeans

# K means iterations
wcss=[]
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10, random_state=0)
    kmeans.fit(male_scaled_data)
    wcss.append(kmeans.inertia_)
# Plot the elbow graph
plt.plot(range(1,11),wcss)
plt.title('Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('inertias')
plt.show()
```

```python
# Female KMeans

# K means iterations
wcss=[]
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter=300,  n_init=10, random_state=0)
    kmeans.fit(female_scaled_data)
    wcss.append(kmeans.inertia_)
    
# Plot the elbow graph
plt.plot(range(1,11),wcss)
plt.title('Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('inertias')
plt.show()
```

```python
# Both male and female data needs 5 clusters
```

```python
# Male

# K means for 5 clusters

kmeans_male = KMeans(n_clusters=5, random_state=0).fit(male_scaled_data)
identified_clusters_male = kmeans_male.predict(male_scaled_data)

# Female

# K means for 5 clusters

kmeans_female = KMeans(n_clusters=5, random_state=0).fit(female_scaled_data)
identified_clusters_female = kmeans_female.predict(female_scaled_data)
```

```python
#Appending the identified clusters to the dataframe for each gender data

# Male
clustered_data_male = male_data.copy()
clustered_data_male['Cluster'] = identified_clusters_male
display(clustered_data_male.head(3))

# Female
clustered_data_female = female_data.copy()
clustered_data_female['Cluster'] = identified_clusters_female
display(clustered_data_female.head(3))
```

```python
fig_male = px.scatter_3d(clustered_data_male, x='Age', y='Annual Income (k$)', z='Spending Score (1-100)',
              color='Cluster')
fig_male.show()
```

```python
fig_female = px.scatter_3d(clustered_data_female, x='Age', y='Annual Income (k$)', z='Spending Score (1-100)',
              color='Cluster')
fig_female.show()
```

```python
# PCA for each data

# Male
pca_male = PCA(n_components=2)
principalComponents_male = pca_male.fit_transform(male_scaled_data)
principalDf_male = pd.DataFrame(data = principalComponents_male
             , columns = ['principal component 1', 'principal component 2'])

principalDf_male['Cluster']  = clustered_data_male['Cluster'].values
display(principalDf_male.head(3))

#Female
pca_female = PCA(n_components=2)
principalComponents_female = pca_female.fit_transform(female_scaled_data)
principalDf_female = pd.DataFrame(data = principalComponents_female
             , columns = ['principal component 1', 'principal component 2'])

principalDf_female['Cluster']  = clustered_data_female['Cluster'].values
display(principalDf_female.head(3))
```

```python
# Male averages
cluster0 = clustered_data_male.loc[clustered_data_male['Cluster']==0]
display(cluster0.mean())

cluster1 = clustered_data_male.loc[clustered_data_male['Cluster']==1]
display(cluster1.mean())

cluster2 = clustered_data_male.loc[clustered_data_male['Cluster']==2]
display(cluster2.mean())

cluster3 = clustered_data_male.loc[clustered_data_male['Cluster']==3]
display(cluster3.mean())

cluster4 = clustered_data_male.loc[clustered_data_male['Cluster']==4]
display(cluster4.mean())
```

```python
# Female averages
cluster0 = clustered_data_female.loc[clustered_data_female['Cluster']==0]
display(cluster0.mean())

cluster1 = clustered_data_female.loc[clustered_data_female['Cluster']==1]
display(cluster1.mean())

cluster2 = clustered_data_female.loc[clustered_data_female['Cluster']==2]
display(cluster2.mean())

cluster3 = clustered_data_female.loc[clustered_data_female['Cluster']==3]
display(cluster3.mean())

cluster4 = clustered_data_female.loc[clustered_data_female['Cluster']==4]
display(cluster4.mean())
```

```python
# Male PCA plot
fig_male = px.scatter(principalComponents_male, x=0, y=1, color=principalDf_male['Cluster'])
fig_male.show()
```

```python
# explained variance male
sum_pca_male = pca_male.explained_variance_ratio_[0] + pca_male.explained_variance_ratio_[1]
sum_pca_male
```

```python
# Female PCA plot
fig_female = px.scatter(principalComponents_female, x=0, y=1, color=principalDf_female['Cluster'])
fig_female.show()
```

```python
# explained variance female
sum_pca_female = pca_female.explained_variance_ratio_[0] + pca_female.explained_variance_ratio_[1]
sum_pca_female
```

```python
# Weighted averaging for male and female
num_of_males = male_data.count()[0]
num_of_females = female_data.count()[0]
all_people = num_of_males + num_of_females

weighted_averaging = (sum_pca_male * num_of_males + sum_pca_female * female_data.count()[0]) / all_people
weighted_averaging
```

```python
# Summary:
# For 2 clusters based on gender the explained variance: 0.667
# For 5 clusters without gender the explained variance: 0.775
# For 5 clusters only male the explained variance: 0.767
# For 5 clusters only female the explained variance: 0.796
# Weighted averaging for male and female = 0.783
```
